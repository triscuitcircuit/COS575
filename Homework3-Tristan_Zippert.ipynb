```python
######################
# Imports
######################
import scipy.io as sio
import numpy as np
```

```python
######################
#Load the data
######################
bodyfat = sio.loadmat("bodyfat_data.mat")
X = bodyfat["X"]
Y = bodyfat["y"]
n,d = X.shape
```

```python
######################
# Train/Test split
######################
X_train, X_test = X[:150:],X[150:]
Y_train, Y_test = Y[:150:],Y[150:]
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

```python
######################
# Bias
######################
bias_1 = np.zeros((64,1)).T
bias_2 = np.zeros((16,1)).T
bias_3 = np.zeros((1,1)).T
```

Set the seed value to 0 and learning rate to 1e-7

```python
np.random.seed(0)
lr = 1e-7
```

```python
######################
# Weights
######################
weights_1 = np.random.randn(64,X_train.shape[1]).T
weights_2 = np.random.randn(16,64).T # previous weight_1 dimension 
weights_3 = np.random.randn(1,16).T  # previous weight_2 dimension
weights_1.shape, weights_2.shape, weights_3.shape
```

### Relu Calculation
The relu calculation is stated as the maximum value with `np.maximum(0,x)`, and the derivative of relu can be calculated with `np.double(x > 0)`.

```python
def relu(x, derivative=False):
    if derivative:
        return np.double(x > 0) 
    else:
        return np.maximum(0,x)
relu(1,True)
```

```python
#Given
def forward_pass(X,W1,W2,W3,b1,b2,b3):
    z1=np.dot(X,W1)+b1
    a1=relu(z1)
    z2=np.dot(a1,W2)+b2
    a2=relu(z2)
    y_hat =np.dot(a2,W3)+b3
    return z1,z2,y_hat,a1,a2

def backward_pass(Xtr, ytr,y_hat,W1,W2,W3,b1,b2,b3,z1,z2,a1,a2,learning_rate):
    #loss function
    loss=np.mean((y_hat-ytr)**2)
    print(loss)
    #backpropogation code
    l3_error=(y_hat-ytr)/float(ytr.shape[0])
    l2_error=l3_error.dot(W3.T)
    l2_delta=l2_error*relu(z2,derivative=True)
    l1_error=l2_delta.dot(W2.T)
    l1_delta=l1_error*relu(z1,derivative=True)
    #update parameters using GD
    W3-= learning_rate*a2.T.dot(l3_error)
    W2-= learning_rate*a1.T.dot(l2_delta)
    W1-= learning_rate*Xtr.T.dot(l1_delta)
    b3-= learning_rate*np.sum(l3_error)
    b2-= learning_rate*np.sum(l2_delta,axis = 0)
    b1-= learning_rate*np.sum(l1_delta,axis = 0)
    return W1,W2,W3,b1,b2,b3, loss
```

The following block of code performs gradient descent by using both the supplied backward and forward pass functions. Stopping threshold is `10000` iterations.

```python
epochs = 10000
for i in range(epochs):
    z1, z2, y_hat, a1, a2 = forward_pass(X_train,
                 weights_1,weights_2,weights_3,
                 bias_1,bias_2,bias_3)
    weights_1, weights_2, weights_3, bias_1, bias_2, bias_3,loss = backward_pass(X_train,
                                                                            Y_train, y_hat, 
                                                                            weights_1, weights_2, weights_3, 
                                                                            bias_1, bias_2, bias_3, 
                                                                            z1, z2, 
                                                                            a1, a2,lr)
```

Perform the forward pass on `X_train` for evaluation. 

```python
z1, z2, y_hat_tr, a1, a2 = forward_pass(X_train,
                 weights_1,weights_2,weights_3,
                 bias_1,bias_2,bias_3)
```

```python
MSE_tr = np.mean((y_hat_tr-Y_train)**2)
f"Training MSE:{MSE_tr}"
```

Perform the forward pass on `X_test` for evaluation. 

```python
z1, z2, y_hat_te, a1, a2 = forward_pass(X_test,
                 weights_1,weights_2,weights_3,
                 bias_1,bias_2,bias_3)
MSE_te = np.mean((y_hat_te-Y_test)**2)
f"Training MSE:{MSE_te}"
```
